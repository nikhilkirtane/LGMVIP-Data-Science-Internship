Image to Pencil Sketch with Python






```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('IMG_20231006_080101.jpg')
```

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
inverted_img = cv2.bitwise_not(gray_img)
```

```python
blurred_img = cv2.GaussianBlur(inverted_img, (111, 111), 0)
```

```python
inverted_blurred_img = cv2.bitwise_not(blurred_img)
```

```python
pencil_sketch = cv2.divide(gray_img, inverted_blurred_img, scale=256.0)
```

```python
plt.figure(figsize=(10, 5))
```

```python
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Actual/Original Image')
plt.axis('off')
```

```python
plt.subplot(1, 2, 2)
plt.imshow(pencil_sketch, cmap='gray')
plt.title('PencilSketch of image')
plt.axis('off')
```

```python
plt.tight_layout()
plt.show()
```
