Data Preparation

```python
sp = {'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3}
```

```python
nk.Classlabels=[sp[i] for i in nk.Classlabels]
```

```python
nk
```

```python
x=nk.iloc[:,0:4]
x
```

```python
y=nk.iloc[:,4]
y
```

Split the data to train and test dataset using train_test_split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

Support Vector Machine Algorithm to
 Train the dataset using svn.fit() method

```python
from sklearn.svm import SVC
svn=SVC()
svn.fit(x_train,y_train)
```

Predict from the test dataset

```python
prediction=svn.predict(x_test)
```

Calculating the Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,prediction)
```
